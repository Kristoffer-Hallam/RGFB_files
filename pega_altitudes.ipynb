```python
code = []
height = []
grav = []
std = []
latitude = []
longitude = []
for folder in folders:
    if string.find(folder, 'DS_Store') != -1:
        continue
    files = os.listdir(current_path+'//estacoes'+'//'+folder)
    for f in files:
        if string.find(f, 'DS_Store') != -1:
            continue
        if string.find(f, '_') != -1:
            continue
        if string.find(f, '.pdf') != -1:
#            print folder, f
            pdf = pdfquery.PDFQuery(current_path+'//estacoes'+'//'+folder+'//'+f)
            pdf.load()
            labelLat = pdf.pq('LTTextLineHorizontal:contains("Latitude")')
            indexLat1 = string.find(labelLat.text(),':')+1
            indexLat2 = string.find(labelLat.text(),'S')-1
            indexLat3 = string.find(labelLat.text(),'N')-1
            if indexLat2 == True:
                latitude.append(string.replace(labelLat.text()[indexLat1:indexLat2], ',', '.'))
            else:
                latitude.append(string.replace(labelLat.text()[indexLat1:indexLat3], ',', '.'))
            labelLon = pdf.pq('LTTextLineHorizontal:contains("Longitude")')
            indexLon1 = string.find(labelLat.text(),':')
            indexLon2 = string.find(labelLat.text(),'O')-1
            longitude.append(string.replace(labelLon.text()[indexLon1:indexLon2], ',', '.'))

            labelAlt = pdf.pq('LTTextLineHorizontal:contains("Altitude geom")')
            indexAlt1 = string.find(labelAlt.text(),':')+1
            indexAlt2 = string.find(labelAlt.text(),' m')
            code.append(f[:-4])
            height.append(string.replace(labelAlt.text()[indexAlt1:indexAlt2], ',', '.'))
            labelGrav = pdf.pq('LTTextLineHorizontal:contains("Gravidade")')
            indexGrav1 = string.find(labelGrav.text(),':')+1
            indexGrav2 = string.find(labelGrav.text(),'m')-1
            grav.append(string.replace(labelGrav.text()[indexGrav1:indexGrav2-7], ',', '.'))
            std.append(string.replace(labelGrav.text()[indexGrav2-5:indexGrav2], ',', '.'))
```

```python
for codei, lati, longi, heighti, gravi, stdi in zip (code, latitude, longitude, height, grav, std):
    print '%s %s %s %s %s %s' % (codei, lati, longi, heighti, gravi, stdi)
```
